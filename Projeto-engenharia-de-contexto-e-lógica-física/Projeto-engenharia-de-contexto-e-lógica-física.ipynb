```python
# PARTE 1 O Algoritmo do Microclima Local

# Dados: [Local, Horário, Temp(°C), Umidade(%), IQA]



def analisar_microclima():

  dados_locais = [

    ["rua Pegmatito",       "17:30l", 25(°C), 66, 2.4],

    ["Avenida Sport Club Corinthians Paulista",       "18:05", 22(°C), 69, 2,5],

    ["Av. Radial Leste, 6745 - Vila Carrão",       "18:45", 20(°C), 74, 2,2],





  cabecalhos = ["Local", "Horário", "Temp (°C)", "Umidade (%)", "IQA"]

  print("=" * 70)

  print("   ANÁLISE DE MICROCLIMA URBANO — SÃO PAULO")

  print("=" * 70)



  for registro in dados_locais:

    local, horario, temp, umidade, iqa = registro

    metricas = [local, horario, temp, umidade, iqa]



    print(f"\n📍 {local} — {horario}")

    print("-" * 50)



    for i, metrica in enumerate(metricas[2:], start=2):

      print(f" {cabecalhos[i]}: {metrica}")



    # --- Classificação por match-case (IQA — faixas oficiais CETESB) ---

    match True:

      case _ if iqa <= 40:

        qualidade = " Boa"

        risco = "Sem risco à saúde."

      case _ if iqa <= 80:

        qualidade = " Moderada"

        risco = "Risco para grupos sensíveis (idosos, crianças, asmáticos)."

      case _ if iqa <= 120:

        qualidade = " Ruim"

        risco = "Risco à saúde da população em geral."

      case _ if iqa <= 200:

        qualidade = " Muito Ruim"

        risco = "Risco sério à saúde. Evitar atividades ao ar livre."

      case _:

        qualidade = " Péssima"

        risco = "Risco gravíssimo. Emergência de saúde pública."



    print(f" Qualidade do Ar: {qualidade}")

    print(f" Risco: {risco}")



    # -------------------------------------------------------------------

    # NOTA DE CONFORTO URBANO (0 a 10)

    #

    # Fórmula:

    # nota_temp  = penaliza temperaturas abaixo de 18°C ou acima de 26°C

    #         usando distância quadrática normalizada

    # nota_umidade = conforto máximo entre 50-70%; penaliza extremos

    # nota_iqa  = escala invertida: IQA 0→10pts, IQA 120+→0pts

    #

    # Nota Final = (nota_temp * 0.35) + (nota_umidade * 0.25) + (nota_iqa * 0.40)

    # -------------------------------------------------------------------



    # Penalidade de temperatura: ideal entre 18-26°C

    temp_ideal_min, temp_ideal_max = 18, 26

    if temp_ideal_min <= temp <= temp_ideal_max:

      nota_temp = 10.0

    else:

      desvio = min(abs(temp - temp_ideal_min), abs(temp - temp_ideal_max))

      nota_temp = max(0, 10 - (desvio ** 1.5))



    # Nota de umidade: ideal entre 50-70%

    umid_ideal_min, umid_ideal_max = 50, 70

    if umid_ideal_min <= umidade <= umid_ideal_max:

      nota_umidade = 10.0

    else:

      desvio_u = min(abs(umidade - umid_ideal_min), abs(umidade - umid_ideal_max))

      nota_umidade = max(0, 10 - (desvio_u * 0.25))



    # Nota IQA: escala invertida linear (0=10pts, 120=0pts)

    nota_iqa = max(0, 10 - (iqa / 12))



    # Nota final ponderada

    nota_conforto = (nota_temp * 0.35) + (nota_umidade * 0.25) + (nota_iqa * 0.40)

    nota_conforto = round(nota_conforto, 2)



    # Classificação verbal da nota

    if nota_conforto >= 8:

      conceito = "Excelente"

    elif nota_conforto >= 6:

      conceito = "Bom"

    elif nota_conforto >= 4:

      conceito = "Regular"

    else:

      conceito = "Desconfortável"



    print(f" Nota de Conforto Urbano: {nota_conforto}/10 — {conceito}")



  print("\n" + "=" * 70)

  print("Análise concluída.")

  print("=" * 70)





analisar_microclima()









PARTE 2 Navegação e Evacuação Espacial

# --- CONFIGURAÇÃO DO MAPA ---

# Lista de locais (nós do trajeto)

locais = [

  "Quarto Master", "Suíte 2", "Banheiro Social", "Corredor",

  "Quarto 3", "Sala de TV", "Sala de Estar", "Cozinha", "Varanda", "Rua"

]



# Lista de estados/obstáculos correspondentes

estados = [

  "Escuro", "Livre", "Vazamento Água", "Fumaça",

  "Chave na Mesa", "Móveis Deslocados", "Livre", "Porta Trancada", "Obstáculo Alto", "Saída"

]



# --- ESTADO DO AGENTE ---

posicao = 0

energia = 15

inventario = []

em_evacuacao = True



print(f"🏠 SIMULADOR DE EVACUAÇÃO ATIVADO")

print(f"O agente inicia no: {locais[posicao]}")

print("-" * 40)



# --- LOOP PRINCIPAL ---

while em_evacuacao and energia > 0:

  cômodo_atual = locais[posicao]

  condicao_atual = estados[posicao]



  print(f"\n📍 Local: {cômodo_atual} | Energia: {energia}")



  # Condição de Vitória

  if condicao_atual == "Saída":

    print("✨ O agente sentiu o ar fresco. Ele chegou à RUA!")

    em_evacuacao = False

    break



  # --- LÓGICA DE NAVEGAÇÃO (IF ANINHADO) ---



  if condicao_atual == "Escuro":

    print("Está tudo escuro. O agente usa as mãos para tatear a parede.")

    posicao += 1 # Avança



  elif condicao_atual == "Vazamento Água":

    print("Chão escorregadio! Risco de queda.")

    energia -= 3 # Perda extra de energia

    posicao += 1



  elif condicao_atual == "Fumaça":

    print("Fumaça densa detectada no corredor!")

    # Subcondição: O agente decide se arrisca ou volta

    if energia > 5:

      print("O agente se abaixa para respirar melhor e prossegue.")

      posicao += 1

    else:

      print("Energia muito baixa para encarar a fumaça. Recuando!")

      posicao -= 1



  elif condicao_atual == "Chave na Mesa":

    print("O agente tateou um objeto: CHAVE COLETADA.")

    inventario.append("chave_saida")

    posicao += 1



  elif condicao_atual == "Porta Trancada":

    print("A porta da Cozinha para a Varanda está trancada!")

    # Subcondição de Inventário

    if "chave_saida" in inventario:

      print("O agente usa a chave que pegou no Quarto 3. Porta aberta!")

      posicao += 1

    else:

      print("Porta bloqueada. O agente precisa retornar para buscar a chave.")

      posicao -= 2 # Retrocede para procurar



  elif condicao_atual == "Obstáculo Alto":

    print("Uma mesa tombada bloqueia a Varanda.")

    if energia > 2:

      print("O agente escala o obstáculo com esforço.")

      energia -= 2

      posicao += 1

    else:

      print("Sem forças para escalar.")

      em_evacuacao = False



  else: # Caso seja "Livre" ou "Móveis Deslocados"

    print("Caminho livre ou com obstáculos leves. Avançando...")

    posicao += 1



  # Redução natural de energia por rodada

  energia -= 1



# --- FIM DA SIMULAÇÃO ---

if energia <= 0:

  print("\n O tempo acabou. O agente ficou exausto antes de sair.")



PARTE 3: Reflexão Crítica

Durante o mapeamento da minha casa, percebi que os ambientes que percorro no "piloto automático" são, na verdade, uma sucessão de eventos. Ao olhar para os meus 3 quartos e o caminho até a varanda, parei de ver apenas móveis e passei a enxergar fluxos de decisão. Notei que a arquitetura do meu lar dita meu comportamento: um corredor com fumaça ou uma porta trancada não são apenas incômodos, mas variáveis críticas que determinam se o objetivo final será alcançado.



O maior desafio foi traduzir o mundo físico para a lógica fria do computador. Criar a condição para a Porta Trancada me mostrou que a solução de um problema na cozinha dependia de uma ação feita lá atrás, no quarto. Isso me ensinou que o progresso nem sempre é uma linha reta; às vezes, a lógica exige dar um passo atrás para coletar o que é necessário e só então avançar. Transformar imprevistos em "custo de energia" me fez entender como é vital quantificar riscos para não desistir no meio do caminho.



No fim das contas, aprender a programar mudou minha forma de encarar o dia a dia. Hoje, em vez de me desesperar com problemas complexos, eu os decomponho em pequenas etapas, como se fossem linhas de código. A programação me deu a clareza de que falhas são apenas erros de parâmetro que podem ser ajustados. Agora, vou passar a olhas para os desafios cotidianos com mais estratégia e paciência, entendendo que, com a lógica certa, sempre existe uma saída, mesmo que eu precise tatear no escuro por um tempo
```
